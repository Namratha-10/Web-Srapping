```python
import requests
import pandas
from bs4 import BeautifulSoup
res=requests.get('https://www.vijaysales.com/tv-and-entertainment/televisions/brand/buy-samsung-televisions')
#print(res)
soup=BeautifulSoup(res.content,'html.parser')
#print(soup)
```

```python
#SAMSUNG
samsung_res=requests.get('https://www.vijaysales.com/tv-and-entertainment/televisions/brand/buy-samsung-televisions')
#print(res)
samsung_soup=BeautifulSoup(samsung_res.content,'html.parser')
#print(soup)
#Product names of Samsung TV
names=samsung_soup.find_all('h2',class_='BcktPrdNm_')
s_names=[]
for i in names:
    a=i.get_text()
    s_names.append(a)
s_names

#Product Prices of Samsung TV
price=samsung_soup.find_all('span',class_='Prdvsprc_')
s_prices=[]
for i in price:
  p=i.get_text()
  s_prices.append(p)
s_prices

#Product EMI of Samsung TV
emi=samsung_soup.find_all('div',class_='BcktPrdemi h54')
s_emi=[]
for i in emi:
   e=i.get_text()
   s_emi.append(e)
s_emi

#Product Exchange of LG TV
exchange=samsung_soup.find_all('div',class_='ofrTag')
s_exchange=[]
for i in exchange:
   x=i.get_text()
   s_exchange.append(x)
s_exchange

#Product URL of Samsung TV
url=samsung_soup.find_all('a',id='pdpId_')
s_url=[]
for i in url:
   u=i['href']
   s_url.append(u)
s_url



















```

```python

lg_res=requests.get('https://www.vijaysales.com/tv-and-entertainment/televisions/brand/buy-lg-televisions')
print(lg_res)

#Product names of LG TV 
lg_soup=BeautifulSoup(lg_res.content,'html.parser')
lg_names=lg_soup.find_all('h2',class_='BcktPrdNm_')
LG_Names=[]
for i in lg_names:
  f=i.get_text()
  LG_Names.append(f)
LG_Names

#Product Prices of LG TV
lg_price=lg_soup.find_all('span',class_='Prdvsprc_')
LG_Price=[]
for i in lg_price:
  f=i.get_text()
  LG_Price.append(f)
LG_Price

#Product EMI Details of LG TV
lg_emi=lg_soup.find_all('div',class_='BcktPrdemi h54')
LG_EMI=[]
for i in lg_emi:
  f=i.get_text()
  LG_EMI.append(f)
LG_EMI

#Product Exchange Details of LG TV
lg_exchange=lg_soup.find_all('div',class_='ofrPrd')
LG_Exchange=[]
for i in lg_exchange:
  f=i.get_text()
  LG_Exchange.append(f)
LG_Exchange

#Product URL og LG TV
lg_url=lg_soup.find_all('a',id='pdpId_')
LG_URL=[]
for i in lg_url:
  f=i['href']
  LG_URL.append(f)
LG_URL
```

```python
# Creating Table from all the Web Scraped Data on Samsung TV
Samsung={'names':s_names,
         'price':s_prices,
         'emi':s_emi,
         'exchange':s_exchange,
         'url':s_url}
Samsung
df=pandas.DataFrame(Samsung)
df
```

```python
# Creating Table from all the Web Scraped Data on LG TV
LG=pandas.DataFrame({'Name':LG_Names,'Price':LG_Price,'EMI':LG_EMI,'EXCHANGE':LG_Exchange,'URL':LG_URL})
LG
```

```python

price=pandas.DataFrame({'Samsung':Samsung['price'],'Lg':LG['Price']})
price
```

```python
# Convert all values to strings, remove currency symbol and commas, then convert to integers
price['Samsung'] = price['Samsung'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '').astype('int')
price['Lg'] = price['Lg'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '').astype('int')
```

```python
# Plotting Bar Chart To represent the difference in the average of Price of Both the TV Brands

import matplotlib.pyplot as plt
plt.bar('Samsung',price['Samsung'].mean())
plt.bar('LG',price['Lg'].mean())
plt.title('Average Price Comparison of Television of Samsung and LG')
plt.plot()

# Average Price Difference between Samsung and LG
avg_difference=price['Samsung'].mean()-price['Lg'].mean()
print('Average Price Difference between Samsung and LG is ',avg_difference)
```
